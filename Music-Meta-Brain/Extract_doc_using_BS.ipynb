```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import re
```

```python
source = urlopen('https://en.wikipedia.org/wiki/Ilaiyaraaja').read()
```

```python
soup = BeautifulSoup(source,'lxml')
print(soup)
```

```python
print(set([text.parent.name for text in soup.find_all(text=True)]))
```

```python
# Extract the plain text content from paragraphs

text = ''
for paragraph in soup.find_all('p'):
  text += paragraph.text

text
```

```python
# Clean text
text = re.sub(r'\[.*?\]+', '', text) # Drop footnote superscripts.
text = text.replace('\n', '') # replace \n with empty string.
text = text.replace('\'s', '')
text
```
